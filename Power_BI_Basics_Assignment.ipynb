6. Define Data View, Report View, and Model View. Explain the purpose of each view.

- Report View
The Report View is the primary canvas where you design the visual interface of your dashboard. This is the space where data becomes a story through interactive charts and maps.
Primary Purpose: Data Storytelling. It allows you to transform complex datasets into digestible visuals like bar charts, slicers, and tables for end-users to consume and interact with.

Data View-
The Data View allows you to inspect the actual "raw" data in your tables after it has been loaded into the model. Think of this as the "Excel spreadsheet" view of your Power BI file.
Primary Purpose: Data Inspection and DAX Creation. It is used to verify data quality, format columns (e.g., setting a column as "Currency"), and create Calculated Columns or Measures using DAX (Data Analysis Expressions).
Model View
The Model View (formerly known as Relationship View) is the "backbone" of your Power BI file, where you manage how different tables connect to one another.
Primary Purpose: Schema Management. It is used to define "Star Schemas," set relationship cardinality (e.g., One-to-Many), and manage cross-filter directions. It ensures that when you click a filter in the Report View, the correct data is filtered across all related tables.

7. Discuss the different data sources that Power BI supports.

- Supported Data Source Categories:
File-Based Sources: Excel, CSV, Text, JSON, XML, PDF, and folders containing multiple files.
Database Sources: SQL Server, Oracle Database, MySQL, PostgreSQL, IBM DB2, Microsoft Access, and others, often supporting DirectQuery for real-time reporting.
Azure Cloud Sources: Azure SQL Database, Azure Synapse Analytics, Azure Data Lake Storage, Azure Cosmos DB, and Azure Databricks.
Online Services & SaaS: SharePoint Online, Salesforce, Google Analytics, Dynamics 365, LinkedIn Sales Navigator, and Jira, enabling direct integration without exporting files.
Other/Advanced Sources: Web pages (via URL), Data feeds, ODBC, Active Directory, Python/R scripts, Spark, and SAP HANA.

8.  Split Owner Name to create two new columns as First Name and Last Name

- Open Power Query Editor: In Power BI Desktop, click on the Transform data button in the Home tab to open the Power Query Editor.
Select the Column: In the Power Query Editor window, select the Owner Name column you want to split.
Split the Column:
Go to the Home tab or the Transform tab on the ribbon.
Click on Split Column.
Select By Delimiter from the dropdown menu.
Specify Delimiter:
In the Split Column by Delimiter dialog box, the delimiter will likely default to Space, which is correct for splitting names.
For "Split at", choose Each occurrence of the delimiter to ensure proper splitting even if there are middle names (though the current data only shows two parts).
Click OK.
Rename Columns: The original "Owner Name" column will now be split into "Owner Name.1" and "Owner Name.2". Double-click on the column headers to rename them to First Name and Last Name respectively.
Apply Changes: Click Close & Apply in the Home tab of the Power Query Editor to load the changes back into your Power BI report.
